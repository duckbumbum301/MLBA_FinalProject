```python
import os
from google.colab import userdata
import google.generativeai as genai
import sys

# In ra phiên bản để kiểm tra
print(f"Phiên bản google-generativeai: {genai.__version__}")

print("\nĐang lấy API key từ 'Secrets'...")
try:
    # 1. Lấy key
    GOOGLE_API_KEY_RAW = userdata.get('GOOGLE_API_KEY')

    # 2. Làm sạch key (cắt bỏ dấu cách/xuống dòng bị dư)
    GOOGLE_API_KEY = GOOGLE_API_KEY_RAW.strip().split('\n')[0]

    print(f"Key đã được làm sạch (kiểm tra): {GOOGLE_API_KEY[:5]}...{GOOGLE_API_KEY[-4:]}")

    # 3. Cấu hình
    genai.configure(api_key=GOOGLE_API_KEY)

    print("Đang khởi tạo model 'gemini-2.5-flash'...")
    model = genai.GenerativeModel('gemini-2.5-flash')

    print("Đang gửi yêu cầu 'Xin chào'...")
    response = model.generate_content("Xin chào, bạn là ai?")

    print("\n--- KẾT QUẢ TỪ GEMINI ---")
    print(response.text)
    print("-----------------------------------")
    print("\nCHÚC MỪNG! BẠN ĐÃ THÀNH CÔNG!")

except Exception as e:
    print(f"\n!!! VẪN BỊ LỖI TRÊN SỔ TAY MỚI !!!")
    print(f"Chi tiết lỗi: {e}")
```
